```python
# Мерджим персоны по неполному совпадению имен

ex1 = 'Журналиста Ивана Сафронова приговорили к 22 годам лишения свободы и Владимир Путин умер и Иван пошел гулять'
ex2 = 'Журналиста Ивана Сафронова приговорили к 22 годам лишения свободы за «государственную измену» — с отбыванием в колонии строгого режима. «Госизмена», по версии следствия, заключалась в том, что для рассылки о России и Восточной Европе он собирал материалы о российской армии — большинство из которых, как следует из обвинительного заключения, можно найти в интернете в открытом доступе. В России статья о «госизмене» сформулирована предельно широко, завести уголовное дело по ней можно даже за СМС-сообщение о передвижении войск, отправленное другу, — а минимальный порог наказания по этой статье (12 лет) выше, чем в советское время за «измену родине» (10 лет). Но даже во всем массиве приговоров «госизменникам» (когда-нибудь невиновные обязательно будут реабилитированы) 22 года, которые дали Ивану, выглядят беспрецедентно жестоко. Сафронов должен отсидеть больше, чем предполагаемый убийца Немцова, один из организаторов захвата «Норд-Оста» и «скопинский маньяк». Как убедительно показали наши коллеги из «Русской службы Би-Би-Си», поводом для этого уголовного дела стала именно журналистская деятельность Ивана. Невозможно отделаться от мысли, что кто-то из героев его материалов хотел свести с Сафроновым личные счеты — это первое (хотя и не единственное) объяснение непомерной жестокости прокуроров и суда. И если это так, то мы, журналисты, обязаны найти этого человека.'

def persons_from_spacy (t):
    doc = nlp(t)
    res = []
    for token in doc.ents:
        if token.label_ == 'PER':
            res.append([token.text, token.start_char, token.end_char])
    return res

def persons_merge_spacy (t):
    from thefuzz import fuzz
    #ner_results = nlp(t)
    person_list = persons_from_spacy(t)


    #for item in ner_results:
        # Проверяем, что длина имени больше 1 символа, иначе там часто лезут пунктуаторы
        #if (item['entity_group'] == 'PER') and (len(item['word']) > 1):
            #person_list.append([item['word'], item['start'], item['end']])
    
    # Упорядочиваем список по длине имени начиная с самых длинных
    person_list.sort(key=lambda x: len(x[0]), reverse=True)

    # Для каждой персоны ищем совпадающее (нестрого) имя в списке с результатом. 
    # Если оно там уже есть, то добавляем к нему новые координаты, если его там нет, то добавляем его самого как новый элемент (с данными координатами).
    # Идем по списку всех найденных персон в порядке убывания длины имени. 
    # Это обеспечивает попадание самого длинного варианта имени в результирующий список (независимо от порядка следования вариантов в тексте). 
    # Этот вариант потом можно использовать как красивый лейбл и для мерджа персон между текстами.
    res = {}
    for item in person_list:
        if len(res) == 0:
            res[item[0]] = [(item[1], item[2])]
        else:
            found = False
            for person in res.keys():
                # partial_ratio пока ставлю больше 80 Для сравнения:
                # 'Maxim Leitsch' vs 'Max Leitsch' = 82; 'Maxim Leitsch' vs 'Reis' = 75; 'Christopher Stark' vs 'Chris Stark' = 73
                if fuzz.partial_ratio(item[0], person) > 75:
                    res[person].append((item[1], item[2]))
                    found = True
                    break
            if not found:
                res[item[0]] = [(item[1], item[2])]
    
    # Сортируем координаты в результатах по порядку следования
    for item in res:
        res[item].sort(key=lambda x: x[0])

    return(res)

print(persons_from_spacy(ex5))
print(persons_merge_spacy(ex5))
```
